```python
import pylab 
import numpy as np
```

## Customizing the plot: Line Styles

```python
pylab.plot(x, y, '-r')  # solid red line ('r' comes from RGB color scheme)
pylab.xlim(0, 10)
pylab.ylim(-1.2, 1.2)

pylab.xlabel('this is x!')
pylab.ylabel('this is y!')
pylab.title('My First Plot')
```

$$ \text{Other options for the color characters are:}  $$ 

$$ \text{  'r' = red }  $$ 
 $$ \text{'g' = green}  $$ 
 $$ \text{'b' = blue}  $$ 
 $$ \text{'c' = cyan}  $$ 
 $$ \text{'m' = magenta}  $$ 
 $$ \text{'y' = yellow}  $$ 
 $$ \text{'k' = black}  $$ 
 $$ \text{'w' = white}  $$ 
 
$$ \text{Options for line styles are}  $$ 

 $$ \text{'-' = solid}  $$ 
 $$ \text{'--' = dashed}  $$ 
 $$ \text{':' = dotted}  $$ 
 $$ \text{'-.' = dot-dashed}  $$ 
 $$ \text{'.' = points}  $$ 
 $$ \text{'o' = filled circles}  $$ 
 $$ \text{'^' = filled triangles}  $$ 
 

### Cusomizing the Plot: Legends

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

pylab.plot(x, y1, '-b', label='sine')
pylab.plot(x, y2, '-r', label='cosine')
pylab.legend(loc='upper right')
pylab.ylim(-1.5, 2.0)
```

### Exercise: Linestyles & Plot Customization

```python
x1 = np.linspace(0, 10, 20)
y1 = np.sin(x1)

x2 = np.linspace(0, 10, 1000)
y2 = np.sin(x2)


pylab.plot(x1, y1, 'bo', label='sampled')
pylab.plot(x2, y2, ':k', label='continuous')
pylab.legend()

pylab.ylim(-1.5, 2.0)
```

## Image Save on current directory

```python
import numpy as np
import matplotlib.pyplot as plt

saffron = np.array([[[245,100,27] for var in range(1920)] for var in range(360)])
white = np.array([[[245,245,245] for var in range(1920)] for var in range(360)])
green = np.array([[[25,107,36] for var in range(1920)] for var in range(360)])
flag = [*saffron,*white,*green]
flag = np.array(flag, dtype=np.uint8)  # change data type for image
plt.imshow(flag)
plt.show()
plt.imsave("India.jpg",flag)
print("Picture is Downloaded on your system")
```
